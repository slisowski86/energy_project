```
#dodajemy kolumny building_type oraz total_square_footage do ramki danych
#tworzymy funkcję mapującą


def to_dict(df, key, value):
  dict=df.rdd
  key_value=dict.map(lambda x: (x[key], x[value]))
  ready_dict=key_value.collectAsMap()
  mapping = create_map([lit(x) for x in chain(* ready_dict.items())])
  return mapping


  
```

```

#sprawdamy działanie funkcji
mapping_metadata=to_dict(df_metadata, 'dataid', 'building_type')
```

```
#dodajemy kolumny
#tworzymy map_type dla building type i total_square_footage
mapping_building=mapping_metadata
mapping_footage=to_dict(df_metadata,'dataid', 'total_square_footage')


```
